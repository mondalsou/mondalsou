# Medicinal Chemistry Alerts

Design and test structural alert rules for EDDSD:

- PAINS-like motifs
- Michael acceptors
- Nitro aromatics, anilines, aldehydes
- Excessively lipophilic aromatic systems

Use RDKit SMARTS queries and explain each rule in markdown.


```python
from rdkit import Chem

patterns = {
    "Michael acceptor": "C=CC=O",
    "Nitro aromatic": "[NX3](=O)=O-[c]",
}

smiles = "O=[N+](=O)c1ccccc1C=CC=O"  # example molecule
mol = Chem.MolFromSmiles(smiles)

for name, smarts in patterns.items():
    patt = Chem.MolFromSmarts(smarts)
    print(name, bool(mol.HasSubstructMatch(patt)))
```
